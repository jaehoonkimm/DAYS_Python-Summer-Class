```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
# print(anscombe)

dataset_1 = anscombe[anscombe['dataset']=='I']
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
print(dataset_1.mean(),dataset_1.std())
print(dataset_2.mean(),dataset_2.std())
print(dataset_3.mean(),dataset_3.std())
print(dataset_4.mean(),dataset_4.std())
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.plot(dataset_1["x"],dataset_1["y"], 'or')
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot(dataset_1['x'],dataset_1['y'], 'bo')
axes2.plot(dataset_2['x'],dataset_2['y'], 'ro')
axes3.plot(dataset_3['x'],dataset_3['y'], 'go')
axes4.plot(dataset_4['x'],dataset_4['y'], 'yo')
```

```python
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

fig.tight_layout()

fig.suptitle("Anscombe Data")
fig
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
fig =plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'], bins=20)

axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Frequency")
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])

axes1.set_title("Scatterplot of Total Bill vs Tip")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=="Female"]['tip'],
              tips[tips['sex']=='Male']['tip']],
            labels = ['Female',"Male"])

#axes1.boxplot([tips[성별이 여성인][tip], tips[성별이남성인][tip]])

axes1.set_xlabel("Sex")
axes1.set_ylabel('Tip')
axes1.set_title("Boxplot of TIps by Sex")
```
