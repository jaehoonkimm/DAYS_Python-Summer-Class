```python
from numpy import NaN, NAN,nan
import pandas as pd
```

결측값이란
0이나 ''은 0이란 값 혹은 빈 문자열이라는 특성을 가지고 있습니다.
하지만 결측값은 아무런 값이 존재하지 않습니다.

비유를 하자면 0이나 ''는 int 혹은 str이라는 박스가 있고, 그 안에 아무것도 없다면
결측값은 그런 박스 조차도 없다고 생각할 수 있습니다.

```python
#아무런 비교할 값이 없기 때문에 NaN == Nan으로 비교하더라도 False가 반환됩니다.
print(0==0)
print(""=="")
print(NaN == NaN)
print('------------------------------------------------------------------------')
#결측치를 확인하기 위해서는 판다스의 isnull이라는 메서드를 사용합니다.
print(pd.isnull(NaN))
print(pd.isnull(''))
print(pd.isnull(0))
print('------------------------------------------------------------------------')
#결측치가 아닌 경우를 확인하려면 notnull이라는 메서드를 사용 할 수 있습니다.
print(pd.notnull(NaN))
print(pd.notnull(''))
print(pd.notnull(0))
```

```python
#누락값 개수 구하기
ebola = pd.read_csv('country_timeseries.csv')

print(ebola.shape) #총 122개 행과 18개의 열을 가지고 있음
print('------------------------------------------------------------------------')
print(ebola.count())
print('------------------------------------------------------------------------')
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count()
print(num_missing)
print('------------------------------------------------------------------------')
print(ebola.Cases_Guinea.value_counts(dropna=False)) #dropna는 결측치를 데이터에 포함하는지 여부
```

```python
#결측치 대체하는법 - 결측치를 특정값으로 대체
# print(ebola.iloc[:,:5].head())
# print('------------------------------------------------------------------------')
# print(ebola.iloc[:,:5].fillna("결측치").head())
print(ebola['Cases_Guinea'])
print(ebola['Cases_Guinea'].mean())
```

```python
#바로 직전 행의 값으로 대체
print(ebola.iloc[:,:5].head())
print('------------------------------------------------------------------------')
print("앞행 값으로 대체")
print(ebola.iloc[:,:5].fillna(method='ffill').head())
print('------------------------------------------------------------------------')
#바로 다음 행의 값으로 대체
print("뒷행 값으로 대체")
print(ebola.iloc[:,:5].fillna(method='bfill').head())
print('------------------------------------------------------------------------')
#앞뒤 값의 평균으로 대체
print("앞뒤값 평균으로 대체")
print(ebola.iloc[:,:5].interpolate().head())
print('------------------------------------------------------------------------')
#결측값 삭제
print("결측값 삭제")
print(ebola.iloc[:,:5].dropna().head())
```
