### [조건]
형식, 내용은 자유이며, 아래의 조건만 모두 충족시켜주시면 됩니다!
- 처음 실행시 : 게임 횟수 input 받기
(예시/ 숫자 5 입력시, 총 5판의 게임을 진행)

- 매 판이 진행될 때마다, computer는 가위, 바위, 보 중에서 하나의 랜덤한 값을 가지고 있음(computer가 어떤 값을 가지고 있는지 플레이어는 알 수 없음)

- 매 판이 진행될 때마다, 사용자는 "가위", "바위", "보" 중에서 하나를 입력해야함.
만약 다른 값을 입력시
(예시/ "가ㅏ위", "123", "ㅇ ", " ", 처럼 예상치 못한 값), 잘못 입력된 값을 무시하고 다시 입력 받기

- 최초에 설정했던 판 수 만큼 게임이 모두 진행되면, 최종 스코어를 판정하여 승, 패, 무승부(+ 각자 스코어)를 알려주기.
(예시/ 총 5판이 진행되어 computer가 3판을 이겼다면, computer 승(3판 승리)을 출력, 
총 10판이 진행되어 computer, 플레이어가 각각 5판을 이겼다면, 무승부(5대5) 출력)

```python
# DAYS 2020 여름 방중활동 3주차 과제 : 가위바위보 게임 만들기

import random

print("가위 바위 보 게임을 시작합니다.")
case_list = ["가위", "바위", "보"]
# 조건 1. 처음 실행 시에 게임 횟수를 input으로 받기
num = int(input("게임 횟수를 입력해주세요 : "))
win_count = 0
lose_count = 0

# 현재 서로의 선택(가위,바위,보)를 나타내주는 함수
def now_choice(user, computer):
    print("- User :", user, "Computer :", computer)

# 승패 여부(user_win)에 따라, win_count 또는 lose_count를 변화시키고, 결과를 출력하는 함수
def win_lose_counter(user_win, user):
    global win_count
    global lose_count
    now_choice(user, computer)
    if user_win==0:
        print("=====> User가 졌습니다! <=====")
        lose_count = lose_count+1
    elif user_win==1:
        print("=====> User가 이겼습니다! <=====")
        win_count = win_count+1
    else:
        print("=====> 비겼습니다! <=====")

# User의 "가위/바위/보"를 parameter로 넣었을 때, Computer와의 승패를 판정해주는 함수
def match_with_computer(user, computer, win_count, lose_count):
    if user==computer:
        user_win = 2 #무승부의 경우
    else:
        if user=="가위":
            if computer=="바위":
                user_win = 0 #유저 패배의 경우
            else:
                user_win = 1 #유저 승리의 경우
                
        elif user=="바위":
            if computer=="보":
                user_win = 0 #유저 패배의 경우
            else:
                user_win = 1 #유저 승리의 경우
                
        elif user=="보":
            if computer=="가위":
                user_win = 0 #유저 패배의 경우
            else:
                user_win = 1 #유저 승리의 경우
    win_lose_counter(user_win, user)
        
        
################# 실질적으로 프로그램이 돌아가는 부분
################# (위에서 def로 정의해준 함수들은 호출 전까지 실행되지 않음)
for case in range(num):
    print("= "*30)
    print("-", case+1,"회차 게임입니다.")
    
    # 조건 2. Computer는 내부적으로 가위, 바위, 보 중 하나의 값을 랜덤하게 가지고 있으며, 이는 User가 알 수 없음.
    computer = random.choice(case_list)
    
    # 조건 3. User는 매 판이 진행될 때마다 가위, 바위, 보 중 하나의 값을 입력해야 하며, 
    # 다른 값을 입력시, 입력된 값을 무시하고 다시 입력 받도록 함
    user_input = input("- 가위 바위 보 중 하나의 값을 입력해주세요 : ")
    # User input이 가위, 바위, 보 중 어느 것도 아니면, 재입력 받음
    while(user_input != "가위" and user_input != "바위" and user_input != "보"):
        user_input = input("- 가위 바위 보 중 하나의 값을 입력해주세요 : ")
        
    match_with_computer(user_input, computer, win_count, lose_count)

    
    
# 조건 4. 처음 설정한 판 수 만큼 게임이 모두 진행되면, 최종 스코어를 판정하여 승/패/무승부와 각자 스코어를 알려주기
print("= "*30)
print("= "*30)
print("<최종 스코어>")
if win_count==lose_count:
    print("User :", win_count, "판 승리")
    print("Computer :", lose_count, "판 승리")
    print("무승부 : ", num-(win_count+lose_count), "판")
    print("최종 결과 : 무승부")
elif win_count>lose_count:
    print("User :", win_count, "판 승리")
    print("Computer :", lose_count, "판 승리")
    print("무승부 : ", num-(win_count+lose_count), "판")
    print("최종 결과 : User의 승리")
elif win_count<lose_count:
    print("User :", win_count, "판 승리")
    print("Computer :", lose_count, "판 승리")
    print("무승부 : ", num-(win_count+lose_count), "판")
    print("최종 결과 : Computer의 승리")
    
    
```
